### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]


#рохраховує мінімальне та максимальне значення
max_n = list_with_random_numbers[0]
min_n = list_with_random_numbers[0]

for i in range(1, len(list_with_random_numbers)):
    if list_with_random_numbers[i] > max_n:
        max_n = list_with_random_numbers[i]
for i in range(1, len(list_with_random_numbers)):
    if list_with_random_numbers[i] < min_n:
        min_n = list_with_random_numbers[i]

        
#середнє значення послідовності        
medium = sum(list_with_random_numbers) / len(list_with_random_numbers)


#медіану послідовності
def median(list_with_random_numbers):
    sortedLst = sorted(list_with_random_numbers)
    lstLen = len(list_with_random_numbers)
    index = (lstLen - 1) // 2
   
    if (lstLen % 2):
        return sortedLst[index]
    else:
        return (sortedLst[index] + sortedLst[index + 1])/2.0
    


import statistics
statistics.median(list_with_random_numbers)
        

print('Max =', max_n)
print('Min =', min_n)
print('Максимальне число =', max(list_with_random_numbers))
print('Мінімальне число =', min(list_with_random_numbers))
print('Medium =', medium)
print('Median(func) =', median(list_with_random_numbers))
print('Median(lib) =', statistics.median(list_with_random_numbers))



                    
```

```python
#частоту кожної цифри в послідовності
dct = {}
def analysis(list_with_random_numbers, dct):
    for i in list_with_random_numbers:
        if i in dct:
            dct[i] += 1
        else:
            dct[i] = 1

analysis(list_with_random_numbers, dct)   

for i in sorted(dct):
    print(i, dct[i])    
```

```python
#середнє квадратичне відхилення

list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

from math import sqrt

lst = list_with_random_numbers

print(sqrt(sum((i-medium)**2 for i in lst)/(len(lst) - 1)))
```
